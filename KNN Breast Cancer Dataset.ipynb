```python
data_df = data_amend.drop(['id','Unnamed: 32'], axis=1)
```

```python
data_df.head(2)
```

```python
data_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data_df)
scaled_features = scaler.transform(data_df)
```

```python
df_feat = pd.DataFrame(scaled_features, columns=data_df.columns)
```

```python
df_feat.head()
```

```python
X = df_feat
y = data_br['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
prediction = knn.predict(X_test)
print(metrics.accuracy_score(y_test,prediction))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(metrics.confusion_matrix(y_test,prediction))
print('___________________________________________')
print(metrics.classification_report(y_test,prediction))
```

```python
"""
Accuracy - overall how often is model correct?
TP + TN / total
"""
164 / 171
```

```python
"""
Misclassification Rate-overall how often is hte model wrong?
FP + FN / total(171)
"""
7/171
```

```python
"""
n=171          Predict 0   Predict 1
Actual 0[[(TN)105    (FP)  3]
Actual 1 [     4     (TP) 59]]

Sensitivity- when actual value is positve, how often is it correct?
   Also known as "recall"
   TP / TP + FN
"""
59 / 63
```

```python
"""
Specificity- when actual value is negative, how often is it correct?
   Also known as "recall"
   TN / TN + FP
"""
105 / 108
```

```python
#plot testing accuracy vs. model complxity

k_range = range(1,26)
scores = []
for k in k_range:
    knn_1 = KNeighborsClassifier(n_neighbors=k)
    knn_1.fit(X_train, y_train)
    y_pred = knn_1.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    
```

```python
plt.plot(k_range, scores)
```
